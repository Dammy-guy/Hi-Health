```python
# Import Module
from tkinter import *
import tkinter as tk
from tkinter import messagebox


# create root window
root = Tk()
# root window title and dimension
root.title("Hi Health")
# Set geometry (widthxheight)
root.geometry('800x600')


#adding menu bar in root window new item in menu bar labelled as 'New' adding more items in the menu bar
menu = Menu(root)
filemenu = Menu(menu)
filemenu.add_command(label='Save')
filemenu.add_command(label='Save As')
filemenu.add_command(label='Open')
filemenu.add_separator()
filemenu.add_command(label='Exit', command=root.quit)
menu.add_cascade(label=' Files ', menu=filemenu)
toolsmenu = Menu(menu)
toolsmenu.add_command(label ='Copy')
toolsmenu.add_command(label ='Cut')
toolsmenu.add_command(label ='Paste')
toolsmenu.add_command(label ='Scan')
menu.add_cascade(label=' Tools ', menu=toolsmenu)
helpmenu = Menu(menu)
helpmenu.add_command(label='Contact us')
menu.add_cascade(label=' Help ', menu=helpmenu)
sharemenu = Menu(menu)
sharemenu.add_command(label=' Share')
menu.add_cascade(label=' Share ', menu=sharemenu)
root.config(menu=menu)

#adding a label to the root window
# Function to submit health data
def submit_data():
    try:
        heart_rate = int(heart_rate_entry.get())
        blood_pressure = blood_pressure_entry.get()
        weight = float(weight_entry.get())
        
        # Process and store the data (you can expand this part)
        data = {
            'Heart Rate': heart_rate,
            'Blood Pressure': blood_pressure,
            'Weight': weight
        }
        
        # Display a message box with the entered data
        messagebox.showinfo("Data Submitted", f"Data: {data}")
        
        # Here you can add code to save the data to a file or database
        
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid data.")
        
# Function to visualize data
def visualize_data():
    # Example data for visualization
    df = pd.DataFrame({
        'Metric': ['Heart Rate', 'Blood Pressure', 'Weight'],
        'Value': [70, 120, 70]  # Replace with actual data
    })
    
    plt.bar(df['Metric'], df['Value'])
    plt.title('Health Metrics')
    plt.ylabel('Value')
    plt.show()
    
# Create input fields
tk.Label(root, text="Heart Rate:").grid(row=0, column=0)
heart_rate_entry = tk.Entry(root)
heart_rate_entry.grid(row=0, column=1)

tk.Label(root, text="Blood Pressure:").grid(row=1, column=0)
blood_pressure_entry = tk.Entry(root)
blood_pressure_entry.grid(row=1, column=1)

tk.Label(root, text="Weight:").grid(row=2, column=0)
weight_entry = tk.Entry(root)
weight_entry.grid(row=2, column=1)


# Survey Form
def open_webpage():
    url = "http://127.0.0.1:5500/Registration%20form.html"  
    webbrowser.open_new(url) 
    
# Create button with click action linked to open_webpage function
button = tk.Button(root, text="Go to Survey Form", command=open_webpage)
button.grid(row=3, column=0, columnspan=2)

# Create buttons
submit_button = tk.Button(root, text="Submit Data", command=submit_data)
submit_button.grid(row=4, column=0, columnspan=2)

visualize_button = tk.Button(root, text="Visualize Data", command=visualize_data)
visualize_button.grid(row=5, column=0, columnspan=2)


#set button grid
#btn.grid(column=2, row=0)


#adding additional Entry by taking "root" as main declaration




# all widgets will be here
# Execute Tkinter
root.mainloop()
```
